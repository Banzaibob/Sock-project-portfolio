```python
import numpy as np
from collections import Counter
%matplotlib inline
import matplotlib.pyplot as plt
from PIL import Image
import math
```

```python
path="CIS_465/html/images/puppy_image.png"
```

```python
image = Image.open(path).convert("L")
arr = np.asarray(image)
```

```python
plt.axis('off')
plt.imshow(arr, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
arr
```

```python
def q3_1(arr,t):
    pic = []
    for row in arr:
        rows = []
        for ele in row:
            if ele < t:
                rows.append(0)
            elif ele >= t:
                rows.append(255)
        pic.append(rows)
    plt.axis('off')
    plt.imshow(pic)
    plt.show()
```

```python
q3_1(arr,70)
```

```python
q3_1(arr,170)
```

```python
def q3_2(arr):
    pic = []
    for row in arr:
        rows = []
        for ele in row:
            if ele < 70:
                rows.append(0)
            elif ele >= 170:
                rows.append(255)
            else:
                rows.append(ele)
        pic.append(rows)
    plt.axis('off')
    plt.imshow(pic)
    plt.show()
```

```python
q3_2(arr)
```

```python
def q3_3(arr):
    c=255/math.log10(1+abs(255))
    pic = []
    for row in arr:
        rows = []
        for ele in row:
            s=c*(math.log10(1+ele))
            rows.append(s)
        pic.append(rows)
    plt.axis('off')
    plt.imshow(pic)
    plt.show()
```

```python
q3_3(arr)
```

```python
def q3_4(arr,n):
    c=255/(255**n)
    pic = []
    for row in arr:
        rows = []
        for ele in row:
            s=c*ele**n
            rows.append(s)
        pic.append(rows)
    plt.axis('off')
    plt.imshow(pic)
    plt.show()
```

```python
q3_4(arr,.5)
```

```python
q3_4(arr,1)
```

```python
q3_4(arr,1.5)
```
