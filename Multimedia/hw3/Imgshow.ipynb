```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2 as cv
import numpy as np
import sys,os
```

## Load Img

```python
img_path="Image-1.bmp"
img = mpimg.imread(img_path)
```

## Split img to color channels

```python
r,g,b = cv.split(img)
```

## Define functions

```python
def convert_to_gray(img):
    return cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
def display_img(img, gray):
    '''
    img = input image or image like array
    gray = bool for displaying a grayscale image
    '''
    plt.axis('off')
    if(gray):
        imgplot = plt.imshow(img, cmap='gray', vmin=0, vmax=255)
    else:
        imgplot = plt.imshow(img)
```

```python
def plot_histo(img):
    hist, bin_edges = np.histogram(gray.flatten(),256,[0,256])
    distrib = hist.cumsum()
    normalized = distrib * float(hist.max()) / distrib.max()
    plt.hist(img.flatten(),256,[0,256], color = 'blue')
    plt.xlim([0,256])
    plt.show()
```

```python
def equalize_histo(img):
    equalized = cv.equalizeHist(img)
    plot_histo(equalized)
```
