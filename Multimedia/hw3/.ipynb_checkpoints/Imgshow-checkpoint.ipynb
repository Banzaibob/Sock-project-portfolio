```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import sys,os
```

```python
img_path="Image-1.bmp"
img = mpimg.imread(img_path)
arr = np.array(img)

def display_img(img, gray):
    '''
    img = input image or image like array
    gray = bool for displaying a grayscale image
    '''
    print(img)
    plt.axis('off')
    if(gray):
        imgplot = plt.imshow(img, cmap='gray', vmin=0, vmax=255)
    else:
        imgplot = plt.imshow(img)
```

```python
def convert_to_gray(img_path):
    img = mpimg.imread(img_path)
    arr = np.array(img)
    return np.average(arr,2)
```
